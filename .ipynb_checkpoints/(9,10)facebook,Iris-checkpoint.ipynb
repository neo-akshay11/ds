```python
import pandas as pd

# ✅ Step 1: Load the dataset with semicolon separator
df = pd.read_csv('/Users/akshay/Desktop/dsbda_practical/DSBDALExam DataSets/DSBDALExam DataSets/dataset_Facebook.csv', sep=';', encoding='latin1')

# ✅ Print basic info
print("Original Dataset Shape:", df.shape)
print("Columns:", df.columns.tolist())

# ✅ Step 2: Create subsets for each post type
photo_posts = df[df['Type'] == 'Photo']
status_posts = df[df['Type'] == 'Status']
video_posts = df[df['Type'] == 'Video'] if 'Video' in df['Type'].unique() else pd.DataFrame()
link_posts = df[df['Type'] == 'Link'] if 'Link' in df['Type'].unique() else pd.DataFrame()

# ✅ Step 3: Merge two subsets (e.g., Photo and Status)
merged_posts = pd.concat([photo_posts, status_posts], ignore_index=True)

# ✅ Step 4: Sort by Page total likes
sorted_df = df.sort_values(by='Page total likes', ascending=False)

# ✅ Step 5: Transpose the dataset
transposed_df = df.T

# ✅ Step 6: Melt data (convert from wide to long format)
melted_df = pd.melt(df,
                    id_vars=['Type', 'Category'],
                    value_vars=['like', 'share', 'comment'],
                    var_name='Interaction Type',
                    value_name='Count')


# ✅ Step 7: Pivot (cast) back to wide format
cast_df = melted_df.pivot_table(index=['Type', 'Category'],
                                columns='Interaction Type',
                                values='Count',
                                aggfunc='sum').reset_index()

# ✅ Optional: Display results
print("\nPhoto Posts Sample:\n", photo_posts.head())
print("\nMerged Posts (Photo + Status):\n", merged_posts.head())
print("\nSorted by Page total likes:\n", sorted_df[['Page total likes', 'Type']].head())
print("\nTransposed Data (first 5 rows):\n", transposed_df.head())
print("\nMelted Data (long format):\n", melted_df.head())
print("\nPivoted Data (wide format):\n", cast_df.head())





```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/Users/akshay/Desktop/dsbda_practical/newdata/iris.csv")

# --- g. Create data subsets for different species ---
setosa = df[df['variety'] == 'Setosa']
versicolor = df[df['variety'] == 'Versicolor']
virginica = df[df['variety'] == 'Virginica']

print("Subset: Setosa")
print(setosa.head())

# --- h. Merge two subsets (Setosa + Versicolor) ---
merged_df = pd.concat([setosa, versicolor])
print("\nMerged DataFrame (Setosa + Versicolor):")
print(merged_df.head())

# --- i. Sort Data by Petal Length ---
sorted_df = df.sort_values(by='petal.length')
print("\nData Sorted by Petal Length:")
print(sorted_df.head())

# --- j. Transposing Data (first 5 rows) ---
transposed_df = df.head().transpose()
print("\nTransposed Data (first 5 rows):")
print(transposed_df)

# --- k. Melting Data to Long Format ---
df = df.reset_index()  # Assign a unique index column
melted_df = pd.melt(df, 
                    id_vars=['index', 'variety'], 
                    value_vars=['sepal.length', 'sepal.width', 'petal.length', 'petal.width'],
                    var_name='feature', 
                    value_name='value')
print("\nMelted Data (Long Format):")
print(melted_df.head())

# --- l. Casting Data to Wide Format (Fixed) ---
wide_df = melted_df.pivot_table(index='index', 
                                 columns='feature', 
                                 values='value')

# Add variety column back using a merge on index
wide_df = wide_df.merge(df[['index', 'variety']], on='index')
print("\nCasted Data (Wide Format):")
print(wide_df.head())
```

```python
import pandas as pd

# Load the dataset with no header
df = pd.read_csv("/Users/akshay/Desktop/dsbda_practical/DSBDALExam DataSets/DSBDALExam DataSets/Iris/Iris.csv", sep=',', header=None)

# Assign column names manually
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']

# Display the first few rows of the dataset
print(df.head())

# g. Create data subsets for different species
setosa_df = df[df['class'] == 'Iris-setosa']
versicolor_df = df[df['class'] == 'Iris-versicolor']
virginica_df = df[df['class'] == 'Iris-virginica']

# Display the shape of each subset
print("Setosa species subset shape:", setosa_df.shape)
print("Versicolor species subset shape:", versicolor_df.shape)
print("Virginica species subset shape:", virginica_df.shape)

# h. Merge two subsets (e.g., Setosa and Versicolor)
merged_df = pd.concat([setosa_df, versicolor_df], ignore_index=True)

# Display the first few rows of the merged DataFrame
print("\nMerged Setosa and Versicolor subsets:")
print(merged_df.head())

# i. Sort data by Petal Length
sorted_df = df.sort_values(by='petal_length')

# Display the first few rows of the sorted DataFrame
print("\nSorted data by Petal Length:")
print(sorted_df.head())

# j. Transposing Data
transposed_df = df.T

# Display the transposed Data
print("\nTransposed Data:")
print(transposed_df.head())

# k. Melting data to long format
melted_df = pd.melt(df, id_vars=['class'], value_vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

# Display the first few rows of the melted DataFrame
print("\nMelted Data to Long Format:")
print(melted_df.head())

# l. Casting data to wide format (using 'class' as the column identifier and 'variable' as the header)
casted_df = melted_df.pivot_table(index=['class'], columns=['variable'], values='value', aggfunc='first')

# Display the casted (wide) format data
print("\nCasted Data to Wide Format:")
print(casted_df.head())
```

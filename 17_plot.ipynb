```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("/Users/akshay/Desktop/dsbda_practical/newdata/heart.csv")

# a. Histogram: To understand the distribution of 'age'
df['age'].plot(kind='hist', bins=10, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()

# b. Dot Plot: To analyze heart rate (thalach) per patient
plt.plot(df.index, df['thalach'], 'o')
plt.title('Dot Plot of Maximum Heart Rate Achieved')
plt.xlabel('Patient Index')
plt.ylabel('Thalach')
plt.show()

# c. Bar Plot: To compare average cholesterol across chest pain types
df.groupby('cp')['chol'].mean().plot(kind='bar', color='skyblue')
plt.title('Avg Cholesterol per Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Cholesterol')
plt.show()

# d. Line Chart: To track resting blood pressure with age
plt.plot(df['age'], df['trestbps'], marker='o')
plt.title('Resting BP vs Age')
plt.xlabel('Age')
plt.ylabel('Resting BP')
plt.show()

# e. Box Plot with Histogram and Scatter
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='chol', data=df)  # Cholesterol by Heart Disease status
sns.stripplot(x='target', y='chol', data=df, color='red', jitter=0.2, size=4)
plt.title('Cholesterol Levels with Scatter over Box Plot')
plt.xlabel('Heart Disease (0=No, 1=Yes)')
plt.ylabel('Cholesterol')
plt.show()
```

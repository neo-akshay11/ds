```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/Users/akshay/Desktop/dsbda_practical/newdata/adult_dataset.csv")

# --- s. Create data subsets for different Country, Sex, Race ---
us_data = df[df['native-country'] == 'United-States']
female_data = df[df['gender'] == 'Female']
white_race_data = df[df['race'] == 'White']

print("\nSubset: US Data")
print(us_data.head())

# --- t. Merge two subsets (Female + White Race) ---
merged_df = pd.concat([female_data, white_race_data]).drop_duplicates()
print("\nMerged Subset (Female + White Race):")
print(merged_df.head())

# --- u. Sort Data using 'hours-per-week' (used as proxy for 'customer ratings') ---
sorted_df = df.sort_values(by='hours-per-week', ascending=False)
print("\nSorted Data by hours-per-week (proxy for ratings):")
print(sorted_df.head())

# --- v. Transposing Data (first 3 rows) ---
transposed_df = df.head(3).transpose()
print("\nTransposed Data:")
print(transposed_df)

# --- w. Melting Data to Long Format ---
df = df.reset_index()  # Give unique index
melted_df = pd.melt(df, 
                    id_vars=['index','income'], 
                    value_vars=['age', 'hours-per-week', 'capital-gain', 'capital-loss'],
                    var_name='feature', 
                    value_name='value')

print("\nMelted Data (Long Format):")
print(melted_df.head())

# --- x. Casting Data to Wide Format ---
wide_df = melted_df.pivot_table(index='index', 
                                 columns='feature', 
                                 values='value')

# Add back income column for context
wide_df = wide_df.merge(df[['index','income']], on='index')
print("\nCasted Data (Wide Format):")
print(wide_df.head())
```

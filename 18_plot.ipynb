```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/Users/akshay/Desktop/dsbda_practical/newdata/heart.csv")

# a. Histogram — To observe distribution of age
plt.hist(df['age'], bins=10, edgecolor='black')
plt.title('Histogram: Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# b. Pie Chart — To show proportion of heart disease cases
counts = df['target'].value_counts()
labels = ['No Disease', 'Disease']
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart: Heart Disease Proportion')
plt.show()

# c. Box Plot — To detect outliers in cholesterol
sns.boxplot(y=df['chol'])
plt.title('Box Plot: Cholesterol Levels')
plt.ylabel('Cholesterol')
plt.show()

# d. Scatter Plot — To see relationship between age and max heart rate
plt.scatter(df['age'], df['thalach'], c='green')
plt.title('Scatter Plot: Age vs Max Heart Rate')
plt.xlabel('Age')
plt.ylabel('Thalach')
plt.show()

# e. Scatter Plot with Boxplots — Add boxplots to scatterplot for age vs cholesterol
sns.jointplot(data=df, x='age', y='chol', kind='scatter', marginal_kws=dict(bins=15, fill=True))
plt.suptitle('Scatter with Marginal Boxplots: Age vs Cholesterol', y=1.02)
plt.show()
```

```python
import pandas as pd

# Load data
df = pd.read_csv('/Users/akshay/Desktop/dsbda_practical/DSBDALExam DataSets/DSBDALExam DataSets/forestfires/forestfires.csv')  # replace with your CSV path

# a. Create data subsets based on area
def classify_area(area):
    if area == 0:
        return 'NotAffected'
    elif area <= 50:
        return 'PartiallyAffected'
    else:
        return 'MostlyAffected'

df['FireImpact'] = df['area'].apply(classify_area)

# Create subsets
not_affected = df[df['FireImpact'] == 'NotAffected']
partially_affected = df[df['FireImpact'] == 'PartiallyAffected']
mostly_affected = df[df['FireImpact'] == 'MostlyAffected']

# b. Merge two subsets: NotAffected and PartiallyAffected
merged_df = pd.concat([not_affected, partially_affected])

# c. Sort data by temp, wind, and area
sorted_df = df.sort_values(by=['temp', 'wind', 'area'])

# d. Transpose data
transposed_df = df.head(5).transpose()  # transposing first 5 rows for readability

# e. Melting data (long format) for a few columns
# e. Melting data (long format) for a few columns
melted_df = pd.melt(df, id_vars=['month', 'day'], value_vars=['temp', 'wind', 'area'])

# f. Casting data back to wide format using pivot_table to handle duplicates
cast_df = melted_df.pivot_table(index=['month', 'day'], columns='variable', values='value', aggfunc='mean').reset_index()
# Display outputs (optional)
print("Subset Classification (first 5 rows):\n", df[['area', 'FireImpact']].head())
print("\nMerged Data Shape:", merged_df.shape)
print("\nSorted by temp, wind, area:\n", sorted_df[['temp', 'wind', 'area']].head())
print("\nTransposed Sample:\n", transposed_df)
print("\nMelted Format:\n", melted_df.head())
print("\nCast Back to Wide Format:\n", cast_df.head())
```

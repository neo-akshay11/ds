```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/Users/akshay/Desktop/dsbda_practical/newdata/adult_dataset.csv')

# f. Histogram — To understand the distribution of ages
plt.hist(df['age'], bins=10, edgecolor='black')
plt.title('Histogram: Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# g. Dot Plot — To visualize education level (numerical) per person
plt.plot(df.index, df['educational-num'], 'o')
plt.title('Dot Plot: Educational Level')
plt.xlabel('Index')
plt.ylabel('Educational-num')
plt.show()

# h. Bar Plot — To compare average hours worked per week across income groups
df.groupby('income')['hours-per-week'].mean().plot(kind='bar', color='skyblue')
plt.title('Bar Plot: Avg Weekly Work Hours by Income')
plt.ylabel('Hours per Week')
plt.xlabel('Income Group')
plt.show()

# i. Line Chart — To track capital gain variation with age
plt.plot(df['age'], df['capital-gain'], marker='o', linestyle='-')
plt.title('Line Chart: Capital Gain vs Age')
plt.xlabel('Age')
plt.ylabel('Capital Gain')
plt.show()

# j. Box Plot with Histogram and Scatter — To analyze capital loss distribution by gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='capital-loss', data=df)
sns.stripplot(x='gender', y='capital-loss', data=df, color='red', jitter=0.2)
plt.title('Box Plot with Scatter: Capital Loss by Gender')
plt.xlabel('Gender')
plt.ylabel('Capital Loss')
plt.show()
```

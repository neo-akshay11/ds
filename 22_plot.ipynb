```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/Users/akshay/Desktop/dsbda_practical/newdata/adult_dataset.csv')

# a. Histogram  To understand how age is distributed across the population
plt.hist(df['age'], bins=10, edgecolor='black')
plt.title('Histogram: Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# b. Pie Chart  To visualize proportion of income groups
income_counts = df['income'].value_counts()
plt.pie(income_counts, labels=income_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart: Income Group Distribution')
plt.axis('equal')
plt.show()

# c. Box Plot  To detect outliers in hours worked per week
sns.boxplot(y='hours-per-week', data=df)
plt.title('Box Plot: Hours Worked per Week')
plt.ylabel('Hours per Week')
plt.show()

# d. Scatter Plot  To analyze relation between age and hours-per-week
plt.scatter(df['age'], df['hours-per-week'], alpha=0.6)
plt.title('Scatter Plot: Age vs Hours Worked Per Week')
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.show()

# e. Scatter Plot with Boxplots  To see distribution of hours worked by gender along with individual points
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='hours-per-week', data=df)
sns.stripplot(x='gender', y='hours-per-week', data=df, color='red', jitter=0.2)
plt.title('Box Plot with Scatter: Hours Worked by Gender')
plt.xlabel('Gender')
plt.ylabel('Hours per Week')
plt.show()
```
